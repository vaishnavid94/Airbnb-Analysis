```python
import numpy as np
import pandas as pd
import sklearn
from sklearn import ensemble
from sklearn import datasets
from sklearn import model_selection
import statsmodels.api as smf

# Split the data into train and test set
X = airbnb.iloc[:, 9:10].values
y = airbnb.iloc[:, 8].values
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.3)
```

```python
# Fitting Simple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the Test set results
y_pred1 = regressor.predict(X_test)
y_pred1

mse_error = np.mean((y_pred1-y_test) ** 2)

print(mse_error)
```

```python
model = sklearn.ensemble.RandomForestRegressor()
result = model.fit(X_train, y_train)
y_pred2 = result.predict(X_test)
mse_error2=np.mean((y_pred2-y_test)**2)
print(mse_error2)
# We used two sklearn techniques to compare the predicted values to predict the pricing of the places.Here it can be observed that, The values of MSE for linear regression and random forest regression are almost same.
```
